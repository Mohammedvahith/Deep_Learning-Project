```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```python
# Load the pre-trained emotion recognition model
model = load_model('emotion_detection_model.h5')  # Update with your model's path

print("Model loaded successfully.")
# Load Haar Cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
# Class labels corresponding to the indices
class_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

def preprocess_face(face_img):
    """Preprocess the face image for prediction."""
    face_img = cv2.resize(face_img, (48, 48))
    face_img = cv2.cvtColor(face_img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    face_img = image.img_to_array(face_img)
    face_img = np.expand_dims(face_img, axis=0)  # Add batch dimension
    face_img = face_img.astype('float32') / 255.0  # Normalize to [0, 1]
    return face_img
```

```python
# Start video capture (0 for webcam or replace with video file path)
cap = cv2.VideoCapture(0)  # Replace with 'video_file_path.mp4' to use a video file

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    if not ret:
        break

    # Convert frame to grayscale for face detection
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5)

    for (x, y, w, h) in faces:
        # Extract the face from the frame
        face_img = frame[y:y+h, x:x+w]

        # Preprocess the face for prediction
        processed_face = preprocess_face(face_img)

        # Predict the emotion
        predictions = model.predict(processed_face)
        predicted_class = np.argmax(predictions, axis=1)[0]
        predicted_label = class_labels[predicted_class]

        # Draw a rectangle around the face and label the emotion
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)  # Draw rectangle
        cv2.putText(frame, predicted_label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)  # Add text

    # Display the resulting frame
    cv2.imshow('Emotion Recognition', frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
